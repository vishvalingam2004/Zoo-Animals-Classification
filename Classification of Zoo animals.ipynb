```python
from sklearn.model_selection import train_test_split
```

```python
x=dr.iloc[:,1:17]
y=dr['class_type']
```

```python
xtr,xts,ytr,yts=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
xtr
```

```python
ytr
```

```python
# ML Model Building.
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
mod1=KNeighborsClassifier()
mod2=DecisionTreeClassifier()
```

```python
mod1.fit(xtr,ytr)
```

```python
mod2.fit(xtr,ytr)
```

```python
pred1=mod1.predict(xts)
print(pred1)
```

```python
pred2=mod2.predict(xts)
print(pred2)
```

```python
# Confusion Matrix.
from sklearn.metrics import confusion_matrix
```

```python
c=confusion_matrix(yts,pred1)
print("Confusion Matrix is: \n",c)
```

```python
sns.heatmap(c,annot=True,fmt='g',cmap='winter',xticklabels=['1','2','3','4','5','6'],yticklabels=['1','2','3','4','5','6'])
plt.xlabel('Actual',fontsize=16)
plt.ylabel('Prediction',fontsize=16)
plt.title("Confusion Matrix",fontsize=20)
plt.show()
```

```python
c=confusion_matrix(yts,pred2)
print("Confusion Matrix is: \n",c)
```

```python
sns.heatmap(c,annot=True,fmt='g',cmap='summer',xticklabels=['1','2','3','4','5','6','7'],yticklabels=['1','2','3','4','5','6','7'])
plt.xlabel('Actual',fontsize=16)
plt.ylabel('Prediction',fontsize=16)
plt.title("Confusion Matrix",fontsize=20)
plt.show()
```

```python
# Evaluating Scores.
from sklearn.metrics import accuracy_score
```

```python
acc=accuracy_score(yts,pred1)
print("Accuracy is:",acc*100)
```

```python
acc=accuracy_score(yts,pred2)
print("Accuracy is:",acc*100)
```
